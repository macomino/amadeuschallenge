```python
s = pd.read_csv('searches_no_dup.csv.bz2', sep='^', usecols=['Destination', 'Origin', 'Date'])
```

```python
s.sample(3)
```

```python
s.reset_index(inplace=True)
s.dropna(inplace=True)
```

```python
s.tail()
```

```python
b.columns=b.columns.map(lambda x:x.strip())
```

```python
b['dep_port'] = b['dep_port'].map(lambda x: x.strip())
```

```python
b['arr_port'] = b['arr_port'].map(lambda x: x.strip())
```

```python
b['cre_date']=b['cre_date'].str[0:10]
```

```python
b['booked'] = 1
```

```python
b.head()
```

```python
b_group = b.groupby(['dep_port','arr_port','cre_date']).sum().reset_index()
```

```python
b_group.head()
```

```python
merge_table = pd.merge(s, b_group, how='left', left_on=['Destination', 'Origin', 'Date'], right_on=['arr_port', 'dep_port', 'cre_date'])
```

```python
merge_table .head()
```

```python
merge_table.drop(['arr_port', 'dep_port', 'cre_date'],axis=1, inplace=True)
```

```python
merge_table.head()
```

```python
merge_table[merge_table['booked'] > 0].head()
```

```python
merge_table[ (merge_table['booked'] > 0)].head()
```

```python
merge_table['booked'].fillna(value=0, inplace=True)
```

```python
merge_table.head()
```

```python
merge_table['booked'] = merge_table['booked'].map(lambda x: 1 if x > 0 else 0)
```

```python
merge_table.head()
```

```python
merge_table.rename(columns={'index':'old_index_from_file'}, inplace=True)
```

```python
merge_table.head()
```
