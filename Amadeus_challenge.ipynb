### Juntamos todos los comandos que sabemos que funcionan para tenerlos en una sola celda

```python
import pandas as pd
b = pd.read_csv('bookings_sample.csv.bz2', sep='^', usecols=['arr_port', 'pax','year'] ,nrows=10000)
b.columns = b.columns.str.strip()
b= b.dropna()
b =b[b['year']==2013]
b = b[['arr_port','pax']]
b = b.groupby('arr_port').sum()
b = b.sort_values('pax', ascending=False)
b = b.head(10)
b
```

```python
bc = pd.read_csv('bookings.csv.bz2', sep='^', usecols=['arr_port', 'pax','year'] , chunksize=1000000)

sum_all_chunks=pd.DataFrame()
for i, chunk in enumerate(bc):
    b= chunk.dropna()
    b =b[b['year']==2013]
    b = b[['arr_port','pax']]
    b = b.groupby('arr_port').sum()
    sum_all_chunks = sum_all_chunks.append(b)

total  = sum_all_chunks.groupby('arr_port').sum()
total = total.sort_values('pax', ascending=False)
total = total.head(10)
total
```
